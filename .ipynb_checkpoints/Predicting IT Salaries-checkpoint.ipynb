```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline
```

# Predicting IT Salaries

This notebook presents an analysis of IT Salary data provided on [Kaggle](https://www.kaggle.com/parulpandey/2020-it-salary-survey-for-eu-region) and derived from an EU Survey in 2019.   

The original data was downloaded from Kaggle and the CSV file was cleaned up in Excel to replace the column headings with names that would be more easily used in Pandas.  

There were 991 individuals represented in the original data.  Some individuals were removed because they had blank fields or represented categories with only one member.  After cleaning there were 825 individuals left in the data. 

Here the data is read from the cleaned CSV file.

```python
salary = pd.read_csv("files/Salary 2019 Cleaned.csv")
print(salary.shape)
salary.head()
```

The columns relating to 2018 salary are incomplete for many individuals with lots of NaN values.  We will make a new column that combines the `Salary`, `Bonus` and `Stocks` columns to represent the total 2019 remuneration for the individual. 

```python
salary['Total'] = salary.Salary + salary.Bonus + salary.Stocks
salary.describe()
```

You can see from the summary here that the salary varies between 6000 and 210000 but the Total remuneration goes from 6000 up to 774000 Euros.  This suggests that at the high salary end, much of the remuneration is in the form of Bonus and Stock payments. We can illustrate this with a plot:

```python
salary.plot(x='Bonus', y='Salary', kind='scatter')
```

Indeed this plot shows that bonus payments are higher for higher salaries up to around 125000 after which they are less common.

# Seniority

We next look at the distribution of individuals by seniority.   The first plot shows the proportion of individuals in each seniority category and shows that the most common levels are `Senior` followed by `Middle`.  

```python
seniority_groups = salary.groupby('Seniority')
seniority_groups.count()['Total'].plot.pie()
```

The next plot shows the mean total salary package for each level. Not surprisingly the `Head` category is paid the most followed by `Principal` and `Lead` as you would expect. 

```python
seniority_groups.mean()['Total'].plot.pie()
```

# Relationships Between Variables

As can be seen from the pairplot below there are some clear relationships between variables in the dataset.  Some of these might be exploited to build a predictive model of total remuneration for an individual.

# Gender

We can look for differences in gender in the data.  The first plot shows the proportion of males and females in the data and clearly has a bias towards males.  The second shows the mean salary for each gender and is more balanced. 

```python
gender_groups = salary.groupby('Gender')
gender_groups.count()['Total'].plot.pie()
```

```python
gender_groups.mean()['Total'].plot.pie()
```

As the next plots show, the age of males and females is equally matched but the experience of females is less than that of males. This might explain why their salary is lower on average. 

```python
gender_groups.mean()['Age'].plot.pie()
```

```python
gender_groups.mean()['Experience'].plot.pie()
```

```python
sns.pairplot(salary[['Age', 'Experience', 'Salary', 'Gender']], hue='Gender')
```

## A Predictive Model

We can build a simple predictive model of Salary using the numeric fields as dependent variables in the regression model.  Here we do this to get a baseline model performance.

```python
salary_r = salary.fillna(0)
```

```python
# create X and y
feature_cols = ['Age', 'Experience', 'Vacation', 'Home']
X = salary_r[feature_cols]
y = salary_r['Total'] 

# import, instantiate, fit
model = LinearRegression()
model.fit(X, y) 
print("y = ", model.coef_, "* x + ", model.intercept_) 
```

```python
mse = ((y - model.predict(X))**2).mean()
print(mse, np.sqrt(mse))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

predicted = model.predict(X)
print("MSE:", mean_squared_error(y, predicted))
print("R^2:", r2_score(y, predicted))
```

The model does not perform well with $R^2$ of 0.19 and a very large MSE.  The next step is to incorporate some of the categorical variable which we have to convert to numerical variables to apply linear regression.   The most obvious one to use is Salary which can be replaced by a number to represent the level of the position.

```python
# create numerical levels for Seniority, provide the ordering of values 
levels = pd.Categorical(salary_r.Seniority, categories=['Junior', 'Middle', 'Senior', 'Lead', 'Principal', 'Head'])
salary_r['Seniority_level'] = levels.codes
salary_r['Gender_level'] = pd.Categorical(salary_r.Gender).codes
```

```python
salary_r['Seniority_level']
```

```python
salary_r['Gender_level']
```

```python
# create X and y
feature_cols = ['Age', 'Experience', 'Vacation', 'Home', 'Seniority_level', 'Gender_level']
X = salary_r[feature_cols]
y = salary_r['Salary'] 

# import, instantiate, fit
model = LinearRegression()
model.fit(X, y) 
print("y = ", model.coef_, "* x + ", model.intercept_) 
```

```python
predicted = model.predict(X)
print("MSE:", mean_squared_error(y, predicted))
print("R^2:", r2_score(y, predicted))
```

The performance of the model has improved a little but is still not very good. There are other categorical variables that might be good predictors of salary such as City, Language and CompanySize.  Including these might give better results.
